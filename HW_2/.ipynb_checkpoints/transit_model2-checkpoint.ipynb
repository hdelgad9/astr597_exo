### Problem 1

In this problem, we want to write a function to find the overlapping area of two circles as a function of their separation $b$ and radius ratio $k = R_p/R_\star$. When the planet is entirely enclosed by the star, the flux merely changes by the square of this ratio $k^2$. However, ingress and egress are more complicated. During the very beginning and the very end of the transit, we will have to solve the geometric problem of finding the overlapping areas of the planet and the star according to the figure below. In this figure, we have actually normalized the distances by the stellar radius $R_\star$ such that $R_\star$ = 1.


To find the overlapping area, we must add the two area sectors and subtract the two triangle areas to give

$$A_\textrm{overlap} = A_1 + A_2 - A_3 - A_4 $$

The area of the first and second chunks, the planet and star sectors, respectively, is given by the typical sector area formula:

$$A_{\textrm{1,2}} = \frac{1}{2}{\Phi}\pi r^2  $$

However, since we are normalizing by the stellar radius $R_\star$, we will have

$$A_{\textrm{1,2}} = {\Phi} \left(\frac{r}{R_\star}\right)^2  $$

With sector angles given by $2\theta$ and $2\phi$, we find

$$A_1 = k^2\phi  $$

and

$$A_2 = \theta $$

The area of the triangle chunks is given by the typical triangle area formulas, where we will have to be clever about how we define the sides and angles to properly assign our free parameters. The law of cosines will be particularly useful in relating the separation $d$ to the angles $\theta$ and $\phi$.

The law of cosines gives

$$k^2 = d^2 + 1^2 - 2d\cos(\theta) $$

$$1^2 = d^2 + k^2 - 2dk\cos(\phi) $$

Solving for $\theta$ and $\phi$ respectively, we find

$$\theta = \cos^{-1}\left(\frac{d^2 + 1 - k^2}{2d}\right) = \cos^{-1}(u)$$

$$\phi = \cos^{-1}\left(\frac{d^2 + k^2 - 1}{2dk}\right) = \cos^{-1}(v)$$

where we have now re-defined $\theta$ and $\phi$ in terms of the free parameters $u$ and $v$.The typical triangle area formula $A = \frac{1}{2}b\times h$ gives

$$A_3 = 2\times\frac{1}{2}\sin(\theta)\cos(\theta) $$

Subbing in our new free parameters gives 

$$A_3 = \sin(\cos^{-1}(v))\cos(\cos^{-1}(v)) = k^2v\sqrt{1-v^2}$$


Similarly, we find for the fourth area chunk

$$A_4 =  2\times\frac{1}{2}k^2\sin(\phi)\cos(\phi)$$ 

$$A_4 = \sin(\cos^{-1}(u))\cos(\cos^{-1}(u)) = u\sqrt{1 - u^2}$$

We can also rewrite our formulae for $A_1$ and $A_2$ in terms of the free parameters $u$ and $v$. Finally, we have a formula for determining the overlapping area of the planet and star as a function of the radius ratio $k$ and the separation $d$ *during ingress and egress*:

$$A_{\textrm{overlap}} =  k^2\cos^{-1}(v) + \cos^{-1}(u) - k^2v\sqrt{1-v^2} - u\sqrt{1 - u^2}$$

```python
'''--------------------------------------------------------------------------'''
'''|                                                                        |'''
'''|                                                                        |'''
'''|                 Import modules and create plot settings                |'''
'''|                                                                        |'''
'''|                                                                        |'''
'''--------------------------------------------------------------------------'''

import pandas as pd
import numpy as np
import math
from scipy.optimize import minimize
import emcee
import corner
from multiprocessing import Pool
from astropy.timeseries import BoxLeastSquares
import astropy.units as u

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import rc
plt.style.use('classic')
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
rc('figure', facecolor='w')
rc('xtick', labelsize=20)
rc('ytick', labelsize=20)

G = 6.674e-8
```

```python
def blockedArea(d, k):
    """
    Compute fraction of blocked area during ingress/egress
    d  : sky separation between planet and star [Rs]
    k  : radius ratio Rp/Rs
    """
    u = (d**2 + 1 - k**2)/(2*d)
    v = (d**2 + k**2 - 1)/(2*d*k)
    
    A1 = k**2 * np.arccos(v)
    A2 = np.arccos(u) 
    A3 = k**2 * v * np.sqrt(1 - v**2)
    A4 = u * np.sqrt(1 - u**2)
    
    return A1 + A2 - A3 - A4
```

### Problem 2

We now wish to apply the overlapping area function we devised in problem 1 in order to write a function for measuring the secondary eclipse in the transit light curve sans limb darkening. We accomplish this by calculating the "dip" in flux due to the planet's transit. We apply three distinct cases: when the planet is wholly outside the star, when the planet is wholly inside it, and when the planet is crossing the star during ingress and egress. In the first case, the flux of the star should not be dimmed at all. In the second case, the flux should be dimmed by the area of the planet. Finally, in the third case, the flux should be dimmed according to our overlapping area equation. In order to apply these cases as a piecewise function, we must write functions to solve for the true longitude angle as a function of time as well as the sky separation.

This piecewise function can be expressed as

$$F = F_0 \begin{cases} 
      1 & d > 1 + k \\
      (1 - (A_1 + A_2 - A_3 - A_4)) & 1 - k \leq d \leq 1 + k \\
      (1 - k^2) & d < 1 - k \\ 
   \end{cases}
$$

The true longitude gives us the ecliptic longitude for $i = 0$ and can be written

$$\theta = \frac{2\pi}{P}(t - t_0) $$

where $P$ is the orbital period, $t$ is our grid of time values, and $t_0$ is the time of inferior conjunction.

Finally, we may express the sky separation as a function of the orbital radius $a$, the true longitude $\theta$, and the inclination $i$:

$$d = a\sqrt{\cos(\theta)^2 + \cos(i)^2\sin(\theta)^2} $$

```python
def longAngle(t, t0, P):
    """
    True longitude angle as a function of time [rad]
    t  : time array [days]
    t0 : time of inferior conjunction [days]
    P  : period [days]
    """
    return 2 * math.pi/ P*(t - t0)


def skySep(theta, a, i):
    """
    Compute projected sky separation btwn planet and star
    theta  : true longitude [rad]
    a      : orbital radius [Rs]
    i      : inclination angle [rad]
    """
    d = a * np.sqrt(np.cos(theta)**2 + np.cos(i)**2 * np.sin(theta)**2)
    return d
```

```python
def transitFlux(tarr, k, a, i, t0, P, F0=1):
    """
    Compute transit light curve as a function of time and orbital parameters
    t  : time array [days]
    k  : radius ratio Rp/Rs
    a  : orbital radius [Rs]
    i  : inclination angle [rad]
    t0 : time of inferior conjunction [days]
    P  : period [days]
    """
    model_flux = [] # Initialize an array for storing the calculated flux during transit.
    
    for t in tarr:
        theta = longAngle(t, t0, P) # Find the true longitude.
        d = skySep(theta, a, i) # Find the sky separation.

        # When the planet is in front of the star:
        if 0 < theta < math.pi:
            if (d > 1 + k): # if the sky separation exceeds 1 + R_p/R_s,
                Ft = F0     # the planet is not blocking the star's light and the star's flux is unchanged.
            elif (d >= 1 - k) & (d <= 1 + k): # if the sky separation exceeds 1 - R_p/R_s or is less than 1 + R_p/R_s,
                Ft = F0*(1 - blockedArea(d,k)/math.pi) # the planet is ingressing/egressing the star and the flux decreases as a function of the overlapping area.
            elif (d < 1 - k): # if the sky separation is less than 1 - R_p/R_s, 
                Ft = F0*(1 - k**2) # the planet is entirely inside the star and the flux decreases as a function of the planet's area.
                
        # When the planet is behind the star:
        else: 
            Ft = F0 # ignore the secondary eclipse so that the transit curve we fit is periodic.
            
        model_flux.append(Ft) # Populate the area with the calculated flux.
    
    return np.array(model_flux)
```

Now that we have written our functions to calculate a transit light curve, we play around with the inclination $i$, the radius ratio $k = \frac{R_p}{R_\star}$, and the orbital radius $a$ to get an intuition for how these parameters affect the shape of the light curve. Playing with these parameters will help to give us good initial guesses for when we attempt to optimize these free parameters.

We first plot the light curve for $k = 0.06, a = 2, t_0 = \frac{\pi}{2}, P = 2\pi$, and for 5 values $i \in \array[1,\frac{\pi}{2}]$. 

```python
tarr = np.linspace(0, 2*math.pi, 500)

# k, a, i, t0, P = .05, .35, math.pi/3, 1, 2*math.pi

plt.figure(figsize=[10,8])
for i in np.linspace(1,math.pi/2,5):
    k = .06
    a = 2
    t0 = math.pi/2
    P = 2*math.pi
    model_flux = transitFlux(tarr, k, a, i, t0, P, F0=1)
    plt.plot(tarr, model_flux, label=r'$i=%.3f$'%(i))
plt.ylim(0.99,1.005)
plt.xlim(0,P)
plt.legend(loc=3)
plt.xlabel('Phase [days]', fontsize=20)
plt.ylabel(r'Normalized Flux', fontsize=20)
plt.ticklabel_format(useOffset=False)
plt.minorticks_on()
plt.show()
```

We find that the inclination nominally affects the width of the light well. A higher inclination (in this case lower $i$) means that the planet merely grazes the top of the star. This gives a narrower transit curve as it would appear that the planet is transitting for less time. In contrast, a lower inclination (higher $i$) means that the planet crosses the star at its widest point, making the planet appear to transit for longer.  We also find that $i = 1$ gives no change in flux. This makes sense, as $i = 0$ implies that the planet never actually occults the star relative to the observer. Similarly, $i = 1$ gives $2\cos(1) = 1.08 > 1 + k$, so the planet's orbit is too highly inclined for $k = 0.06$ for us to observe a reduction in flux.

We now plot the light curve for $k = 0.06, i = 1.1, t_0 = \frac{\pi}{2}, P = 2\pi$, and for 5 values $a \in \array[1,2.2]$. 

```python
plt.figure(figsize=[10,8])
for a in np.linspace(1.2,2.2,5):
    k = .06
    i = 1.1
    t0 = math.pi/2
    P = 2*math.pi
    model_flux = transitFlux(tarr, k, a, i, t0, P, F0=1)
    plt.plot(tarr, model_flux, label=r'$a=%.3f$'%(a))
plt.ylim(0.99,1.005)
plt.xlim(0,P)
plt.legend(loc=3)
plt.xlabel('Phase [days]', fontsize=20)
plt.ylabel(r'Normalized Flux', fontsize=20)
plt.ticklabel_format(useOffset=False)
plt.minorticks_on()
plt.show()
```

We find that $a$ again scales with the transit width. If we pick a large enough $a$, it also impacts the transit depth. This makes sense, as the farther the planet gets from the star, the smaller the star appears relative to the planet. As a result, we receive a smaller fraction of light from the star. Below this boundary, as the planet gets closer to the star, the transit width increases. We can apply the same logic used to explain the trends in inclination to explain this effect.

Finally, we plot the light curve for $i = 1.1, a = 2, t_0 = \frac{\pi}{2}, P = 2\pi$, and for 5 values $k \in \array[0,0.1]$. 

```python
plt.figure(figsize=[10,8])
for k in np.linspace(0,.1,5):
    i = 1.1
    a = 2
    t0 = math.pi/2
    P = 2*math.pi
    model_flux = transitFlux(tarr, k, a, i, t0, P, F0=1)
    plt.plot(tarr, model_flux, label=r'$k=%.3f$'%(k))
plt.ylim(0.99,1.005)
plt.xlim(0,P)
plt.legend(loc=3)
plt.xlabel('Phase [days]', fontsize=20)
plt.ylabel(r'Normalized Flux', fontsize=20)
plt.ticklabel_format(useOffset=False)
plt.minorticks_on()
plt.show()
```

As expected, we find that $k$ scales with the trnasit depth, where a higher $k$ gives a deeper transit curve. Higher $k$ means that the radius of the planet $R_p$ is increasing relatively to the stellar radius $R_\star$. The larger the planet, the more of the stellar flux it will block out, resulting in a deeper transit curve.

We can attribute some initial guesses to these parameters with a combination of observations and assumptions. First, when we plot the curve, we observe that the minimum occurs at $0.996F_0$. Thus, the transit depth appears to be $\sim 0.004$. Since the depth gives the ratio $k^2$, we can guess that $k \approx \sqrt{0.004} = 0.06$. We initially guess that the orbit is near edge-on, giving $i < \pi/2 \approx 1.56$. We must guess $i < \pi/2$ so that we do not find an impact parameter $b = 0$. Finally, the width of the light well will give us the approximate orbital radius. From our plot, we guess $a \approx 1.5\:R_\star$. 

### Problem 3

We now want to use our model to fit the data ‘mystery_planet02.txt’ and find the optimized period $P$, depth $k$, impact parameter $b$, duration of the transiting body $t_0$, and the density of the star $\rho_\star$.


```python
'''--------------------------------------------------------------------------'''
'''|                                                                        |'''
'''|                                                                        |'''
'''|                   Import the mystery planet data.                      |'''
'''|                                                                        |'''
'''|                                                                        |'''
'''--------------------------------------------------------------------------'''

data = pd.read_csv('files/mystery_planet02.txt', delim_whitespace=True, header=None)

time, flux, flux_err = np.array(data[0]), np.array(data[1]), np.array(data[2])
```

First we plot the flux vs. time data from the 'mystery_planet02.txt' file:

```python
plt.figure(figsize=[10,8])
plt.errorbar(time, flux, yerr=flux_err, fmt='.', color='k')
plt.ticklabel_format(useOffset=False)
plt.xlabel('Time [days]', fontsize=20)
plt.ylabel(r'Normalized Flux', fontsize=20)
plt.show()
```

We then need a function that calculates the phase to find the best way to fold the data in terms of the most probable period whcih will be calculated in the next cell. 

```python
def phase_fold(time, flux, flux_err, P):
    
    fold = time % P
    phase_sort = np.argsort(fold)

    phase_time = fold[phase_sort]
    phase_flux = flux[phase_sort]
    phase_flux_err = flux_err[phase_sort]
    
    return phase_time, phase_flux, phase_flux_err
```

By using *astropy* BoxLeastSqaures, an algorithm  used for detecting transiting exoplanets and eclipsing binaries in time series photometric data, we can easily calculate the most probable period. We use the *power* function of the BoxLeastSquares class which calculates the log likelihood of the model fit and we give it a period of up to 50 days.

```python
model = BoxLeastSquares(time * u.day, flux, dy=flux_err)

periods = np.linspace(1, 50, 100000) * u.day
periodogram = model.power(periods, 0.2)
```

We then plot our results from the BoxLeastSquares algorithm and we set the arranged values of time and flux in terms of the phase of the orbit.

```python
best_period = periodogram.period[np.argmax(periodogram.power)].value

phase_time, phase_flux, phase_flux_err = phase_fold(time, flux, flux_err, best_period)

plt.plot(periodogram.period, periodogram.power);
plt.xlabel('Period [days]', fontsize=20);
plt.ylabel("power", fontsize =20)
print('best period:', best_period)
```

From the graph, we can confirm that the best period is at ~9.79 days.

The following function does the following:

1. From given values of $k$, $a$, $i$, $t_0$, it computes the light curve formed by the tranisitng planet over a given period using the *transitFlux* function from Problem 2. 
2. Calculates $\chi^2$ from the modeled light curve and the folded flux from the data.
3. Calculates our other parameters of interest (i.e. $b$, $\delta$, $T$, $\tau$, $\rho$).
4. Plots the light curve fit over the data.

```python
def plot_lc_model(theta, data, P, **kwargs):
    """
    Plot light curve with parameters theta, compute chi^2
    and calculate other parameters of interest (b, delta, T, tau, rho)
    
    theta  : array of transit model parameters (k, a, i, t0)
    data   : phase folded data arrays (phase_time, phase_flux, phase_flux_err)
    P      : best fit period (using BLS)
    """
    k, a, i, t0 = theta
    phase_time, phase_flux, phase_flux_err = data
    
    # compute flux over one period to plot
    tarr = np.linspace(0, P, 1000)
    plot_model_flux = transitFlux(tarr, k, a, i, t0, P, F0=1)
    
    # compute chi^2 of model with parameters given in theta
    model_flux = transitFlux(phase_time, k, a, i, t0, P, F0=1)
    chi_val = np.sum(((phase_flux - model_flux)/phase_flux_err)**2)
    
    # calculate other parameters of interest
    b     = a*np.cos(i)
    delta = k**2 
    T     = np.sqrt(1 - b**2)/(math.pi*a)*P
    tau   = k*T/(1 - b**2)
    T_sec = T * 24 * 3600
    P_sec = P * 24 * 3600
    rho   = 3/math.pi**2 * (P_sec)/(G*(T_sec)**3) * (1 - b**2)**(3/2)
    
    # the following are the values needed to calculate rho based as done on Seager & Mallén-Ornelas (2003)
    #G     = 6.674e-8
    #t_T   = P / math.pi * math.asin((((1 + k)**2 - (a * math.cos(i))**2) / (1 - (math.cos(i))**2))**(1 / 2) / a)
    #t_F   = P / math.pi * math.asin(((1 - k)**2 - (a * math.cos(i))**2)**(1 / 2))
    #P_sec = P * 24 * 3600 # in seconds
    #rho   = 4 * (math.pi)**2 / ((P_sec)**2 * G) * (((1 + math.sqrt(delta))**2 - (b)**2 \
    #          * (1 - (math.sin(t_T * math.pi / P_sec))**2)) / (math.sin(t_T * math.pi / P_sec))**2)**(3/2)
    
    lbl = r'Model fit: $k=%.3f, a=%.3f, i=%.3f, t_0=%.3f, P=%.3f$'%(k, a, i, t0, P)
    lbl2 = r'$b=%.3f, \delta=%.4f, T=%.3f, \tau=%.3f, \rho=%.3f$'%(b, delta, T, tau, rho)
    
    # plot data and model
    plt.figure(figsize=[20,8])
    plt.errorbar(phase_time, phase_flux, yerr=phase_flux_err, fmt='.', color='k', elinewidth=.3)
    plt.plot(tarr, plot_model_flux, color='r', linewidth=2, label=r'$\chi^2=%.3f$'%(chi_val))
    if 'xlim' in kwargs:
        plt.xlim(kwargs.get('xlim'))
    else:
        plt.xlim(0, P)
        plt.text(.15, 1.013, lbl, ha='left', fontsize=18)
        plt.text(.15, .987, lbl2, ha='left', va='top', fontsize=18)
    plt.title('Transit Model Fit', fontsize=22)
    plt.xlabel('Phase [days]', fontsize=20)
    plt.ylabel(r'Normalized Flux', fontsize=20)
    plt.legend(loc='lower right')
    plt.ticklabel_format(useOffset=False)
    plt.minorticks_on()
    if 'save' in kwargs:
        plt.savefig(kwargs.get('save'))
    plt.show()
    
    return b, delta, T, tau, rho
```

```python
k = .06
i = 1.56
a = 15
t0 = 3.36
P = best_period

theta = [k, a, i, t0]
data = phase_time, phase_flux, phase_flux_err

plot_lc_model(theta, data, P, save='transit_model_initial.png');
```

We can see that the function works well and gives a fit that follows the behavior of the data. However, we still need to make sure that the fit and the calculated parameters are the best fit and most probable from the data. We do this by using Baye's Theorem, therefore, we need to define functions to calculate the log-likelihood of our model given the initial values we used for the parameters as the prior.

```python
def log_likelihood(theta, data, P):
    """
    Assume Gaussian likelihood function
    theta : model parameters (k, a, i, t0, P)
    data  : phase-folded time, flux, flux_err
    P     : period, fixed during optimization
    """
    k, a, i, t0 = theta
    phase_time, phase_flux, phase_flux_err = data
    
    phase_flux_model = transitFlux(phase_time, k, a, i, t0, P)
    
    return -0.5 * np.sum((phase_flux - phase_flux_model)**2/phase_flux_err**2)


def log_prior(theta):
    """
    Assume uniform prior for all model parameters k, a, i, t0
    """
    k, a, i, t0 = theta
    if (bounds[0][0] < k < bounds[0][1]) & (bounds[1][0] < a < bounds[1][1]) \
        & (bounds[2][0] < i < bounds[2][1]) & (bounds[3][0] < t0 < bounds[3][1]) & (a*np.cos(i) < 1):
        return 0.0
    return -np.inf


def log_probability(theta, data, P):
    """
    Mulitply prior and likelihood
    """
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, data, P)
```

After calculating the log-probability of the model, from *scipy.optimize* we use the minimize function with the 'nelder-mead' method to find the best values for our parameters and in turn, this will also give the other parameters of interest.

```python
nlp = lambda *args: -log_probability(*args)

initial = [k, a, i, t0]
bounds  = [(0,1), (1,200), (0,math.pi/2), (0,P)]
data    = [phase_time, phase_flux, phase_flux_err]

soln = minimize(nlp, initial, method='nelder-mead', args=(data, P))
```
